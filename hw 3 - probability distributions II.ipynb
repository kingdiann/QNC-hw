```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 178        # number of trials
p0 = 0.5       # chance probability (null)
k_obs = 130     # observed correct responses

# Create distribution
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p0)

# Plot
plt.bar(x, pmf, alpha=0.6, label='Binomial PMF (chance)')
plt.axvline(k_obs, color='red', linestyle='--', label=f'Observed = {k_obs}')
plt.xlabel('Number of correct responses')
plt.ylabel('Probability')
plt.title(f'Binomial distribution (n={n}, p={p0})')
plt.legend()
plt.show()

# p-value: probability of >= observed correct responses
p_value = 1 - binom.cdf(k_obs - 1, n, p0)
print("p-value =", p_value)
```

This code is attempting to replicate figure 3 from the paper: Adibi et al. 2012. Behavioral study of whisker-mediated vibration sensation in rats. Proc Natl Acad Sci U S A 
In this figure they are using nose poke as a measurement of the mouse's ability to discriminate between two different amplitude of vibration - so either nose poke for discrimination or no nose poke for no discrimination. This is a binomial distribution with a total number of trials (n) and number of successful nose pokes (k). Here I am setting p as 0.5 to represent the null probability (i.e. that the mouse is unable to discriminate and the nose poke is random chance). Since this generates a p-value well below 0.05 we can conclude that the results obtained (number of successful discriminations) are not due to chance.
